```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sample Indian Demographics Data (Replace with real Census data if available)
data = {
    'state': ['Maharashtra', 'Uttar Pradesh', 'Bihar', 'West Bengal', 'Tamil Nadu',
              'Madhya Pradesh', 'Rajasthan', 'Karnataka', 'Gujarat', 'Andhra Pradesh'],
    'population': [12314423, 237827725, 124799926, 99609303, 77841267,
                   85387263, 83902869, 67546896, 6457546, 59309931],
    'literacy_rate': [82.9, 73.0, 63.8, 77.1, 80.3, 70.6, 67.5, 73.5, 79.6, 67.4],
    'rural_population_percent': [54.8, 77.7, 88.7, 68.1, 51.7, 72.4, 75.1, 61.3, 57.4, 66.5]
}

# Convert dictionary to Pandas DataFrame
demographics_df = pd.DataFrame(data)

# Display first few rows of the dataset
print(demographics_df.head())

# Display dataset information
print(demographics_df.info())

# Summary statistics
print(demographics_df.describe())

# Check for missing values
print(demographics_df.isnull().sum())

# ----------------------------- DATA VISUALIZATION -----------------------------

#  **Bar Chart: Literacy Rate by State**
plt.figure(figsize=(10,6))
sns.barplot(x="literacy_rate", y="state", data=demographics_df, palette="coolwarm")
plt.xlabel("Literacy Rate (%)")
plt.ylabel("State")
plt.title("Literacy Rate Across Different Indian States")
plt.show()

#  **Scatter Plot: Literacy Rate vs Rural Population Percentage**
plt.figure(figsize=(8,5))
sns.scatterplot(x=demographics_df["literacy_rate"], y=demographics_df["rural_population_percent"], hue=demographics_df["state"], palette="viridis", s=100)
plt.xlabel("Literacy Rate (%)")
plt.ylabel("Rural Population Percentage (%)")
plt.title("Correlation Between Literacy Rate & Rural Population")
plt.show()

#  **Choropleth Map (Optional, requires geoJSON data)**
# If you want to use a map, use plotly with Indian state boundary data

# Save processed dataset as CSV (optional for sharing)
demographics_df.to_csv("indian_demographics.csv", index=False)

print(" Project completed successfully! ")
```
